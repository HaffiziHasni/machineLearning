```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
df=pd.read_csv('new.csv',encoding='latin-1')
print(df)
joined=' '.join(df.cleaned_definition)
```

```python
#building word cloud
wordcloud = WordCloud().generate(joined)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
background_color: "white"
max_words: 50
```

```python
#separating between misogynistic and non
mis=df.loc[df['is_misogyny'] == 1]
non=df.loc[df['is_misogyny'] == 0]
joinedmis=' '.join(mis.cleaned_definition)

#wordcloud for misogyny
wordcloudmis = WordCloud().generate(joinedmis)

plt.imshow(wordcloudmis, interpolation='bilinear')
plt.axis("off")
plt.show()
background_color: "white"
max_words: 50
```

```python
#wordcloud for non misogyny
joinednon=' '.join(non.cleaned_definition)

wordcloudnon = WordCloud().generate(joinednon)

plt.imshow(wordcloudnon, interpolation='bilinear')
plt.axis("off")
plt.show()
background_color: "white"
max_words: 50
```
